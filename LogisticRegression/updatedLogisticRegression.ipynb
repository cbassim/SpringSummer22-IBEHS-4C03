```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

To be able to use head circumference to predict the likelihood that a baby will have brain hemorrhage (Germinal matrix hemorrhage = grmhrm), we fit the logistic regression model.

```python
babies=pd.read_csv('Babies1.csv')
babies
```

```python
babies.info()
```

```python
grmhrmD=pd.get_dummies(babies["grmhrm"],prefix="grmhrmD",drop_first=True)
grmhrmD
```

```python
babies=pd.concat([babies,grmhrmD],axis=1)
babies
```

```python
babies.info()
```

Visualize the data

```python
# Probability of outcome

sns.regplot(x="headcirc",y="grmhrmD_Yes",data=babies,logistic=True)
```

```python
from statsmodels.formula.api import logit
import statsmodels.api as sm
```

```python
model1_logreg=logit("grmhrmD~headcirc",data=babies)
results1=model1_logreg.fit()
print(results1.summary())
```

Coefficent for headcirc=-0.1117:  The log odds of hemorrhage decresaes by -0.117 for every 1 cm increase in head circimference.  

This is not significant based on the z-test, with a p-value above 0.05 and a 95%CI that crosses 0.

After fitting the logistic regression, you will want to calculate the Odds Ratios of the estimated parameters.  Everything you need is available from the Resullts object that comes from the model fit.  You can take the estimated parameters and confidence intervals and apply NumPy's exp() function.

```python
np.exp(results1.params)
```

The estimated odds ratio of a baby having hemmorrhage assocaited with 1 cm increase in head circumfrence is 0.89.  The odds of having hemmorhange / the odds of not having hemmorrahge associated with 1 cm increase in head circumfrenece = less than 1.  This means that increasing head circumfrence is associated with lower odds of having a hemmorrage (not statistcially significnat!). 

Finding predictions using the log odds formula:  logit(p) = log odds(p) = ln[p/(1-p)) = a+bx

```python
predcitionsA=results1.predict(babies)
predcitionsA
```

What is the predicted probability that a baby will have a brain hemmorhage with a head circumfrence of 28cm?

```python
predictions=results1.predict(exog=dict(headcirc=28))
predictions
```
